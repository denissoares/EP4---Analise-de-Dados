# Média de nota por ano de conclusão do ensino médio - Enem 2018


```python
import pandas as pd
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

Lendo CSV para DataFrame

```python
filename = 'MICRODADOS_ENEM_2018.csv'
df = pd.read_csv(filename, sep = ';',encoding = 'ISO-8859-1')
```

```python
df.shape
```

```python
df.describe()
```

```python
df.sample(10)
```

Separando os campo que serão usados

```python
df_notas = df.loc[:,['TP_ANO_CONCLUIU','NU_NOTA_CN' ,'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO']]
```

```python
df_notas.sample(10)
```

Execução do cálculo da média geral

```python
df_notas['MEDIA'] = (df_notas['NU_NOTA_CN'] + df_notas['NU_NOTA_CH'] +
df_notas['NU_NOTA_LC'] + df_notas['NU_NOTA_MT'] + df_notas['NU_NOTA_REDACAO'])/5
```

```python
df_notas.sample(10)
```

Converter código do ano, para ano específico 

```python
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 0, 'ANO_CONCLUSAO_TRATADO'] = 'Não Informado'
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 1, 'ANO_CONCLUSAO_TRATADO'] = 2017
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 2, 'ANO_CONCLUSAO_TRATADO'] = 2016
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 3, 'ANO_CONCLUSAO_TRATADO'] = 2015
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 4, 'ANO_CONCLUSAO_TRATADO'] = 2014
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 5, 'ANO_CONCLUSAO_TRATADO'] = 2013
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 6, 'ANO_CONCLUSAO_TRATADO'] = 2012
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 7, 'ANO_CONCLUSAO_TRATADO'] = 2011
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 8, 'ANO_CONCLUSAO_TRATADO'] = 2010
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 9, 'ANO_CONCLUSAO_TRATADO'] = 2009
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 10, 'ANO_CONCLUSAO_TRATADO'] = 2008
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 11, 'ANO_CONCLUSAO_TRATADO'] = 2007
df_notas.loc [df_notas['TP_ANO_CONCLUIU'] == 12, 'ANO_CONCLUSAO_TRATADO'] = 'Antes de 2007'
```

```python
df_notas.sample(10)
```

Agrupamento dos dados por ano e ordenando pela maior média

```python
df_notas.groupby('ANO_CONCLUSAO_TRATADO').mean().sort_values('MEDIA', ascending=False)
```

Separando colunas para geração de gráfico

```python
df_notas_media = df_notas.loc[:,['ANO_CONCLUSAO_TRATADO','MEDIA']]
```

```python
df_notas_media.groupby('ANO_CONCLUSAO_TRATADO').mean().sort_values('MEDIA').plot(kind='barh', figsize=(10,10))
```

```python
df_notas_media.groupby('ANO_CONCLUSAO_TRATADO').mean().sort_values('MEDIA', ascending=False).plot(kind='bar', y = 'MEDIA',figsize=(10,5))
```

### Conclusão:

Quando mais recente a formação do candidato tem a tendência de maior nota.

# Eliminados em prova - ENEM 2018

```python
df_eliminados = df.loc[:,['SG_UF_RESIDENCIA', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']]
```

```python
df_eliminados.sample(10)
```

Separação dos candidatos que foram eliminados

```python
df_eliminados.query('TP_PRESENCA_CN == 2 or TP_PRESENCA_CH == 2 or TP_PRESENCA_LC == 2 or TP_PRESENCA_MT == 2').sample(10)
```

```python
df_eliminados.query('TP_PRESENCA_CN == 2 or TP_PRESENCA_CH == 2 or TP_PRESENCA_LC == 2 or TP_PRESENCA_MT == 2').groupby('SG_UF_RESIDENCIA')['TP_PRESENCA_CN'].count()
```

```python
df_eliminados.query('TP_PRESENCA_CN == 2 or TP_PRESENCA_CH == 2 or TP_PRESENCA_LC == 2 or TP_PRESENCA_MT == 2').groupby('SG_UF_RESIDENCIA')['TP_PRESENCA_CN'].count().plot(kind='bar',figsize=(10,5))
```

```python
df_eliminados.groupby('SG_UF_RESIDENCIA')['TP_PRESENCA_CN'].count()
```

```python
df_candidatos = df_eliminados.groupby('SG_UF_RESIDENCIA')['TP_PRESENCA_CN'].count()
df_candidatos_eliminados = df_eliminados.query('TP_PRESENCA_CN == 2 or TP_PRESENCA_CH == 2 or TP_PRESENCA_LC == 2 or TP_PRESENCA_MT == 2').groupby('SG_UF_RESIDENCIA')['TP_PRESENCA_CN'].count()
```

```python
df_unificado = pd.merge(df_candidatos, df_candidatos_eliminados, on = 'SG_UF_RESIDENCIA')
```

```python
df_unificado.rename(columns={'SG_UF_RESIDENCIA': 'ESTADO',
                   'TP_PRESENCA_CN_x': 'TOTAL_CANDIDATOS',
                   'TP_PRESENCA_CN_y': 'CANDIDADOS_DESCLASSIFICADOS'
                  }, inplace=True)

df_unificado.sample(27)
```

```python
df_unificado_tratado = df_unificado['PERCENTUAL_DESCLASSIFICADO'] = 100 - (df_unificado['TOTAL_CANDIDATOS'] 
                                              - df_unificado['CANDIDADOS_DESCLASSIFICADOS']) / df_unificado['TOTAL_CANDIDATOS'] * 100
df_unificado.sample(27).sort_values('PERCENTUAL_DESCLASSIFICADO', ascending=False)
```

```python
df_unificado_tratado.sample(27)
```

```python
df_unificado_tratado.plot(kind='bar',figsize=(10,5))
```

### Conclusão:

Não foi o Estado com mais candidatos que foi a maior desclassificação.
